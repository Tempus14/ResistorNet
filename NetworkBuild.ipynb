```python
import networkx as nx
```

```python
G=nx.hypercube_graph(3)
nx.draw_networkx(G)
```

```python
G.nodes()
```

```python
X=G.nodes()
```

```python
F=nx.Graph()
```

```python
F.add_nodes_from(X)
```

```python
nx.draw_networkx(F)
```

```python
F.adj
```

```python
for u in F.nodes():
    F.nodes[u]["pos"]=u
```

```python
F.add_edges_from(nx.geometric_edges(F, 1))
```

```python
F.edges[(0,0,0),(0,0,1)]["weight"]=14
```

```python
F.adj
```

```python
nx.draw_networkx(F)
```
