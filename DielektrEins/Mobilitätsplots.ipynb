```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
holes=[10, 100, 300, 500, 1000]
ten=np.loadtxt("10/mobility.dat.txt", delimiter= " ")
hun=np.loadtxt("100/mobility.dat.txt", delimiter= " ")
thun=np.loadtxt("300/mobility.dat.txt", delimiter= " ")
fhun=np.loadtxt("500/mobility.dat.txt", delimiter= " ")
tau=np.loadtxt("1000/mobility.dat.txt", delimiter= " ")
```

```python
D1=pd.DataFrame(ten)
D2=pd.DataFrame(hun)
D3=pd.DataFrame(thun)
D4=pd.DataFrame(fhun)
D5=pd.DataFrame(tau)
D=pd.concat([D1, D2, D3, D4, D5])
```

```python
plt.plot(D1[0], D1[1], label="10")
plt.plot(D2[0], D2[1], label="100")
plt.plot(D3[0], D3[1], label="300")
plt.plot(D4[0], D4[1], label="500")
plt.plot(D5[0], D5[1], label="1000")
plt.xlabel('Feldstärke')
plt.ylabel("Mobilität")
plt.legend()
plt.show()
```

```python
plt.errorbar(holes, D[1][0], D[2][0], label="0.005")
plt.errorbar(holes, D[1][1], D[2][1], label="0.007")
plt.errorbar(holes, D[1][2], D[2][2], label="0.009")
plt.xlabel('Number of holes')
plt.ylabel("Mobility")
plt.title("Dielektrikum Eins")
plt.legend(title="Field strength")
plt.savefig('mobility.png')
plt.show()
```
