```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import colors as mcolors
import seaborn as sns
```

```python
D=pd.read_table("TEST.xyz", sep=" ",names=["type","x","y","z"], usecols=["x","y","z"], skip_blank_lines=1).dropna(axis=0).reset_index(drop=True)
D.head()
```

```python
holes=10
step=1
colors=list(mcolors.CSS4_COLORS.keys())
fig = plt.figure()    
ax = fig.add_subplot(111, projection='3d')
for i in range(len(colors)):
    offset=i*holes*step
    F=D[offset:offset+holes]


    ax.scatter(F["x"], F["y"], F["z"], c= colors[i] , marker='o')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
plt.show()
```

```python
def PlotGrid(hole, step=5):
    D=pd.read_table(str(hole)+".xyz", sep=" ",names=["type","x","y","z"], usecols=["x","y","z"], skip_blank_lines=1).dropna(axis=0).reset_index(drop=True)
    colors=list(mcolors.TABLEAU_COLORS.keys())
    fig = plt.figure()    
    ax = fig.add_subplot(111, projection='3d')
    for i in range(len(colors)):
        offset=i*hole*step
        F=D[offset:offset+hole]


        ax.scatter(F["x"], F["y"], F["z"], c= colors[i] , marker='o')
        ax.set_xlabel('X-axis')
        ax.set_ylabel('Y-axis')
        ax.set_zlabel('Z-axis')
    plt.savefig("fig/"+str(hole)+".png")
    plt.show()
```

```python
holes=[10,100,300,500,1000]
for j in holes:
    PlotGrid(j,10)
```

```python
holes=[10,100,300,500,1000]
for j in range(5):
    PlotGrid(1000, j)
```

```python
D=pd.read_table("1000.xyz", sep=" ",names=["type","x","y","z"], usecols=["x","y","z"], skip_blank_lines=1).dropna(axis=0).reset_index(drop=True)
holes=1000
step=2
colors=list(mcolors.CSS4_COLORS.keys())
for i in range(3):
    offset=i*holes*step
    F=D[offset:offset+holes]

    ax = sns.heatmap((F["x"], F["z"]))
    plt.show()
```
